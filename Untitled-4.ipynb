```python
#Crearemos una serie de pandas,con los siguientes numeros 25,17,34,82,124,75,44,63,47,9,81,124,17,124, a esta serie la llamaremos numvip, clonaremos la serie para tener 
solamente los numeros mayores de 40 numvip40, los mayores de 40 en orden descendente y reindexando, a esta nueva serie le aplicaremos una funcion que sume 10 y divide por dos a 
cada elemento de la serie, ver en un texto el promedio de la lista numvip40 redondeado a 3 decimales, ver en otro texto la desviacion tipica y en otro texto la varianza.
```

```python
#Practica 1
import pandas as pd
numeros=[25,17,34,82,124,75,44,63,47,9,81,124,17,124]
numvip=pd.Series(data=numeros)
numvip40=numvip[numvip>40].sort_values(ascending=False,ignore_index=True)
numvip40
def suma10d2(x):
    return (x+10)/2
numvip40=numvip40.apply(suma10d2)
numvip40
print("el promedio de NUMVIP40 es",round(numvip40.mean(),3))



```

```python
#Ver la cantidad de veces, que esta repetido cada elemento
numvip40.value_counts()
```

```python
#Como añadir un campo o columna nueva a un dataframe, 3 maneras
#1.Poniendo un valor manual fijo
#2.Seria aplicandole una funcion con Apply
#3.Partiendo de una columna o campo existente
```

```python
#Escribiendo un nombre que exista en un DataFrame
clase["Colegio"]="IT Academy"
clase
```

```python
#Aplicando una funcion, funcion para que en una columna nueva nos ponga ok a los alumnos que midan mas de 1.70

def mide170(x):
    if x>1.70:
        return "ok"
    else:
        return "no ok"
    clase["tipo_1.70"]=clase["Estatura"].apply(mide170)
    clase
```

```python
persona_mayor=clase["Edad"].max()
print(persona_mayor)
clase["Diferencia"]= clase["Edad"].max()-clase["Edad"]
clase
```

```python
#Como eliminar una columna, con la instruccion drop + el parametro axis
#axis = 1 significa columna y axis=0 significa eliminar fila

clase=clase.drop(["MAX"],axis=1)
clase
```

```python
#Eliminar el registro cuyo indice es 2

clase=clase.drop([2],axis=0,inplace=True)
clase
```
